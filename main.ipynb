```python
import numpy as np
import matplotlib.pyplot as plt

# Energy levels and colors
Es = [0.5, 1.0, 2.0]
colors = ['blue', 'green', 'red']
t = np.linspace(0, 2*np.pi, 400)

# Choose five time steps
indices = np.linspace(0, len(t)-1, 5, dtype=int)

fig, axes = plt.subplots(1, 5, figsize=(20, 4))
for ax, idx in zip(axes, indices):
    ax.set_xlim(-2.5, 2.5)
    ax.set_ylim(-2.5, 2.5)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title(f't = {t[idx]:.2f}')
    ax.set_aspect('equal')
    # vector field
    X, Y = np.meshgrid(np.linspace(-2.5,2.5,10), np.linspace(-2.5,2.5,10))
    ax.streamplot(X, Y, Y, -X, color='lightgray', density=1.2)
    for E, c in zip(Es, colors):
        r = np.sqrt(2*E)
        x = r*np.cos(t)
        y = r*np.sin(t)
        ax.plot(x, y, color=c, alpha=0.3)
        ax.plot([y[idx]], [x[idx]], 'o', color=c, label=f"E={E}")
    if idx == indices[0]:
        ax.legend()
plt.tight_layout()
plt.show()
```
